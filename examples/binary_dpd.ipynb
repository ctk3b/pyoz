```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
%matplotlib inline
import numpy as np

import pyoz as oz
```

```python
dpd_binary = oz.System()
r = dpd_binary.r

dpd_binary.set_interaction(0, 0, oz.dpd(r, a=15))
dpd_binary.set_interaction(0, 1, oz.dpd(r, a=17))
dpd_binary.set_interaction(1, 1, oz.dpd(r, a=15))

U_r = dpd_binary.U_r
```

```python
plt.plot(r, U_r[0, 0], marker='D', markevery=10, label='0-0')
plt.plot(r, U_r[0, 1], marker='o', markevery=10, label='0-1')
plt.plot(r, U_r[1, 1], marker='x', markevery=10, label='1-1')
plt.xlabel('r')
plt.ylabel('U(r)')
plt.xlim(0, 1.5)
plt.legend()
```

```python
g_r, c_r, e_r, h_k = dpd_binary.solve(rhos=[1, 2], closure_name='hnc')
```

```python
plt.plot(r, g_r[0, 0], marker='D', markevery=10, label='0-0')
plt.plot(r, g_r[0, 1], marker='o', markevery=10, label='0-1')
plt.plot(r, g_r[1, 1], marker='x', markevery=10, label='1-1')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(0, 2)
plt.legend(loc='best')
```
