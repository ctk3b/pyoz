```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
%matplotlib inline
import numpy as np

import pyoz as oz
```

```python
dpd_unary = oz.System()
r = dpd_unary.r
dpd_unary.set_interaction(0, 0, oz.dpd(r, a=50))
U_r = dpd_unary.U_r
```

```python
plt.plot(r, U_r[0, 0])
plt.xlabel('r')
plt.ylabel('U(r)')
plt.xlim(0, 2)
```

```python
g_r, c_r, e_r, h_k = dpd_unary.solve(rhos=5, closure_name='hnc')
```

```python
plt.plot(r, g_r[0, 0])
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(0, 4)
```
